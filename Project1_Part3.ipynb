## 1.) Importing cleaned dataset, pandas, and matplotlib
#### Dataset: Non Major Survey Data from 2024


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "Non-Majors Survey Results - Fall 2024.csv"
df = pd.read_csv('/Users/faitth/Desktop/CMP-262/Project #1/Project #1 Part 2 - Felicia Hangge/CLEANED_Non-Majors Survey Results - Fall 2024.csv')
```

***
## 2.) Answering Assigned Data Science Question
#### Question #19: Which computing areas has the highest interest for those students interested in taking more computing classes?


```python
# setting plot style
plt.style.use('ggplot')

# assigning relevant columns for easier analysis 
interest_columns = [
    'interest_cybersecurity','interest_app_dev','interest_data_analytics','interest_ML_AI','interest_comp_prog','interest_game_design','interest_hardware' ]

#counting responses for "yes" in each category 
interest_counts = df[interest_columns].apply(lambda x: (x == 'Yes').sum())

#creating plot 
plt.bar(interest_counts.index, interest_counts.values, color = 'indianred')
plt.xlabel("Computing Area")
plt.ylabel("Number of Interested Students")
plt.title("Student Interest in Computing Areas")
plt.xticks(rotation=40, ha ='right')
plt.show
```

***
## 3.) Answering First Personal Data Science Question
#### Question #1: Which program has the most non-major students taking computing courses?

```python
# setting plot style 
plt.style.use('ggplot')

#counting counts for each major 
major_counts = df['major'].value_counts()

#creating plot 
plt.bar(major_counts.index, major_counts.values, color = 'darkseagreen')
plt.xlabel("Major Program")
plt.ylabel("Number of Non-Majors")
plt.title("Most Common Major Programs Among Non-Major Students Enrolled in Computing Courses")
plt.xticks(rotation=45, ha = 'right')
plt.show
```

***
## 3.) Answering Second Personal Data Science Question
#### Question #2: What computing courses are non-majors the most interested in taking in the future?

```python
# setting plot style 
plt.style.use('ggplot')

#counting counts for each course 
future_int_counts = df[interest_columns].apply(lambda x: (x == 'Yes').sum())

#creating plot 
plt.bar(future_int_counts.index, future_int_counts.values, color = 'powderblue')
plt.xlabel("Computing Courses")
plt.ylabel("Number of Interested Non-Majors")
plt.title("Future Computing Course Interests for Non-Majors")
plt.xticks(rotation=45, ha = 'right')
plt.show
```

***
## 4.) Answering Third Personal Data Science Question
#### Question #3: Which age group of non-major students is the most interested in taking more computing courses?

```python
# setting plot style 
plt.style.use('ggplot')

#counting counts of 'yes' in any interest column and grouping by age 
df['interests'] = df[interest_columns].apply(lambda x: (x == 'Yes').any(), axis=1)
age_counts = df[df['interests']].groupby('age').size()

#creating plot 
plt.bar(age_counts.index, age_counts.values, color = 'plum')
plt.xlabel("Age Group")
plt.ylabel("Number of Interested Non-Major Students")
plt.title("Interest in Future Computing Courses by Age Group")
plt.xticks(rotation=45, ha = 'right')
plt.show
```
